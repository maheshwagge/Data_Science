# Matplotlib Basics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0, 10)
```

```python
y = 2 * x
```

```python
x, y
```

```python
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.xlim(0, 10)
plt.ylim(0, 18)
plt.title('simple linear line')

plt.savefig('Figures/sample-fig.png')
```
