# Matplotlib Sub Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
x = np.arange(0, 10)
y = x * 2
```

# Adding rows and columns

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x, y)
axes[1][1].plot(a, b)
```

# Loop through axes

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)
```

------

# fixing overlapping area betweens figures
- plt.tight_layout()
- fig.tight_layout()

```python
fig, axes = plt.subplots(nrows=3, ncols=1)

for ax in axes:
    ax.plot(x, y)

plt.tight_layout()
```

------

# Manually set spacing between figures
* left = 0.125  # the left side of the subplots of the figure
* right = 0.9   # the right side of the subplots of the figure
* bottom = 0.1  # the bottom of the subplots of the figure
* top = 0.9     # the top of the subplots of the figure
* wspace = 0.2  # the amount of width reserved for space between subplots,
              # expressed as a fraction of the average axis width
* hspace = 0.2  # the amount of height reserved for space between subplots,
              # expressed as a fraction of the average axis height

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
axes[1][1].plot(a,b)

# manully set the spacing
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.3,)

plt.show()
```

-------------

# Parameters on subplots()

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8), dpi=100)

axes[0][0].plot(x,y)
axes[0][0].set_title('0 0 Title')

axes[0][1].plot(a,b)
axes[0][1].set_ylabel('ylabel 0,1')
axes[0][1].set_xlim(1, 8)

axes[1][0].plot(x,y)

axes[1][1].plot(a,b)
axes[1][1].set_title('Title 1,1')

# super title
fig.suptitle('Whole Figure Title', fontsize=16)

plt.tight_layout()
plt.show()

fig.savefig('Figures/new_subplots.png', bbox_inches='tight')
```
