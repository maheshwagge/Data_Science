```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 11)
```

# Legends
- add label to each one
- call legend()

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x**2, label='X Vs X^2')

ax.legend(loc="lower left");
```

# Lots of options....

- ax.legend(loc=1) # upper right corner
- ax.legend(loc=2) # upper left corner
- ax.legend(loc=3) # lower left corner
- ax.legend(loc=4) # lower right corner
-...
# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location


```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x**2, label='X Vs X^2')

ax.legend(loc=(1.1, 0.5)); # now the legend is outside as x value is over 1 
```

# Setting colors, linewidths, linetypes
- ### possible linestype options `‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’`

```python
x = np.linspace(0, 11, 10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, color='#e803fc', label='X vs X', linewidth=1.5, linestyle='--') #RGB Hexcode

ax.plot(x, x/10, color='#03fc28', label='X vs X/10', lw=5, ls='-.')

plt.legend()
```

# Custom linestyle dash

```python
fig, ax = plt.subplots(figsize=(10,6))
lines = ax.plot(x,x**2, color='purple', ls='--', lw=2)
lines[0].set_dashes([5, 2, 10, 2]) # 5 points of dash, 2 points of whitespace, 10 points of dash, 2 points of whitespaces
```

# Markers
- https://matplotlib.org/3.2.2/api/markers_api.html
- ax.plot(x, x-1,marker='+',markersize=20)
- ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
- ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
- ax.plot(x, x-4,marker='1',ms=20)

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, marker='o', markersize=10, color='purple', lw=2)
ax.plot(x, x**4, marker='+', markersize=10, color='green', lw=2)
```

# Custom marker edges, thickness,size,and style

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2,color='purple', lw=2,  marker='s', markersize=10, 
       markerfacecolor='red', markeredgecolor='yellow', markeredgewidth=2)
```
